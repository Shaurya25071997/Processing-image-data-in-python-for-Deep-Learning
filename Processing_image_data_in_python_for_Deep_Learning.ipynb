```python
#importing the image module from matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt

```

```python
#loading image through matplotlib.image module

img = mpimg.imread('/content/golden-retriever-tongue-out.jpg')
```

```python
type(img)
```

```python
print(img.shape)
```

```python
print(img)
```

```python
#display the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

**Resizing the image uisng pillow library**

```python
from PIL import Image
```

```python
img = Image.open('/content/golden-retriever-tongue-out.jpg')
```

```python
img_resized = img.resize((200,200))
```

```python
print(img_resized.size)
```

```python
img_resized.save('/content/resized_image.jpg')
```

```python
#display the image from numpy array

img_res = mpimg.imread('/content/resized_image.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**RGB to Grayscale conversion**

```python
#Conerting RGB images to Grayscale image using OpenCV

import cv2
img = cv2.imread('/content/golden-retriever-tongue-out.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
grayscale_image.shape
```

```python
type(grayscale_image)
```

```python
#saving the grayscale image

cv2.imwrite('/content/grayscale_image.jpg', grayscale_image)
```
